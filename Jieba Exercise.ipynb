```python
with open('./cai_speech.txt') as f:
    speech = f.read()
```

```python
wd_count = Counter()
for word in jieba.cut(speech):
    print(word)
    if word in wd_count:
        wd_count[word] += 1
    else:
        wd_count[word] = 1
```
